```python
import itertools

def split_dict(d, n):
    i = iter(d.items())

    d1 = dict(itertools.islice(i, n))
    d2 = dict(i)

    return d1, d2
```

```python
class LeafNet(nn.Module):

    def __init__(self):
        super(LeafNet, self).__init__()

        self.convs = nn.Sequential(
            nn.Conv2d(1, 64, 5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 96, 3),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
        )

        self.lins = nn.Sequential(
            nn.Linear(96 * 13 * 13, (96 * 13 * 13 / 2)),
            nn.Softmax(dim=1),
            nn.Linear((96 * 13 * 13 / 2), (96 * 13 * 13 / 4)),
            nn.Softmax(dim=1),
        )
        
        self.classifier = nn.Linear(4056, 100)
        self.dropout = nn.Dropout(0.15)

    def forward(self, x):
        x = self.convs(x)
        x = self.dropout(x)

        print('shape', x.shape)
        x = x.reshape(x.size(0), -1)
        x = self.lins(x)

        x = self.classifier(x)

        return x


model = LeafNet()
model.to('cuda:0')
```

```python
optimizer = optim.Adam(model.parameters(), lr=3e-4, betas=(0.9, 0.999), eps=1e-7, amsgrad=False) # 3e-4 0.0003

EPOCHS = 1
BATCH = 16

loss_fn = nn.CrossEntropyLoss()
```

```python
(train_d, test_d) = split_dict(data, 1200)
```

```python
# TODO: clean up
def flatten(d):
    images = []
    labels = []
    for (label, arr) in d.items():
        for image in arr:
            images.append(image)
            labels.append(label)
    return images, labels

train = flatten(train_d)
test = flatten(test_d)
```

```python
losses = []
accuracies = []

for i in tqdm(range(EPOCHS)):
    optimizer.zero_grad()
    
    X = torch.tensor().float().to('cuda:0')
    
    Y = torch.tensor().float().to('cuda:0')

    out = model(X)
    
    loss = loss_fn(out, Y)

    losses.append(loss)
    accuracies.append(accuracy)

    loss.backward()

    optimizer.step()
```

```python
for i in range(EPOCHS):
    losses[i] = losses[i].cpu().detach().numpy()
    accuracies[i] = accuracies[i].cpu().detach().numpy()
    
plt.plot(range(EPOCHS), accuracies, 'g', label='Accuracy')
plt.plot(range(EPOCHS), losses, 'b', label='Loss')
plt.title('Accuracy and Loss')
plt.show()
```
